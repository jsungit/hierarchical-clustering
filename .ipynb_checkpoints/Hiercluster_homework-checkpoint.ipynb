```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import pdist, squareform

from scipy.cluster.hierarchy import linkage, dendrogram, fcluster, ward
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
wine = pd.read_csv('data/wine.data', header =None)
wine.columns = ['Group', 'Alcohol', 'MalicAcid', 'Ash', 'Alcalinity', 'Magnesium', 'TotalPhenols', 'Flavanoids', 'NonflavanoidPhenols', 'Proanthocyanins', 'ColorIntensity', 'Hue', ' OD280-OD315ofDilutedWines', 'Proline']
#wine.info()
#wine.head(3)
```

```python
new = wine.copy() 
target = np.array(new.pop("Group"))
#Cluster.size
data = new
```

```python
# scale the data
scalar = StandardScaler().fit(data)
data = scalar.transform(data)
```

```python
#Find the distance between pairs of all 13-dimensional vectors
distxy = squareform(pdist(data, metric='euclidean'))
```

```python
# use linkage to create the clusters
Clusters = linkage(distxy, 'ward')
fig = plt.figure(figsize = (10, 5))
R = dendrogram(Clusters)
```

```python
Clusters = linkage(distxy, 'single')
fig = plt.figure(figsize = (10, 5))
R = dendrogram(Clusters)
plt.show
```

```python
L = ward(linkage(distxy))
C = fcluster(L, t=3, criterion='distance')
```

```python
fcluster(L, t=1000, criterion='distance')
```
